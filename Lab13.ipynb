```python
target_genre = "Comedy"

target_model = md1_df
```

```python
# Precision = tp / (tp + fp)

tp = target_model[target_model["predicted"] == target_genre]["correct?"].sum()
print(tp)

fp = target_model[target_model["predicted"] == target_genre].shape[0] - tp
print(fp)

if (tp + fp) == 0:
    precision = 0
else:
    precision = tp / (tp + fp)
print("Precision:", precision)
```

```python
# recall = tp / (tp + fn)

target_preds = target_model[target_model["actual genres"].str.contains(target_genre)]
fn = target_preds[target_preds["predicted"] != target_genre].shape[0]
print(fn)
```

```python
recall = tp / (tp + fn)
print("Recall:", recall)
```

```python
tn = target_model.shape[0] -tp - fp - fn
print(tn)
accuracy = (tp + tn) / target_model.shape[0]
print(accuracy)
```

LOOPS

```python
models = {"Model 1": md1_df, "Model 2": md2_df, "Model 3": md3_df}
for model_name, target_model in models.items():
    print()
    print(model_name)
    for target_genre in ["Drama", "Comedy", "Romance", "Horror"]:

          tp = target_model[target_model["predicted"] == target_genre]["correct?"].sum()
          fp = target_model[target_model["predicted"] == target_genre].shape[0] - tp

          if (tp + fp) == 0:
            precision = 0
          else:
            precision = tp / (tp + fp)

          target_preds = target_model[target_model["actual genres"].str.contains(target_genre)]
          fn = target_preds[target_preds["predicted"] != target_genre].shape[0]

          recall = tp / (tp + fn)

          if (precision + recall) == 0:
              f1 = 0
          else:
              f1 = (2* precision * recall) / (precision + recall)

          tn = target_model.shape[0] -tp - fp - fn
          accuracy = (tp + tn) / target_model.shape[0]
          print(f"{target_genre} - Accuracy:, {accuracy:.3f} - Percision:{precision:.3f}, Recall:{recall:.3f}, F1:{f1:.3f}")


print("****"* 20)
```
